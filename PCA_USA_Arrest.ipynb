```python
x=scale(df.iloc[:,0:].values)
print x
```

```python
x=np.cov(x.T)
print x
```

```python
eig_val,eig_vec=np.linalg.eig(x.T)
print "eigen vectors \n%s"%eig_vec
print "eigen values \n%s"%eig_val
```

```python
#so in this we can see that e1 ,e2 are two eigen values can explain our data very much 
#so we have two take first two vectors for our analysis 
pca=PCA()
pca.fit_transform(scale(df))
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
#taking 1st component as murder with 62% , 2nd will be Assualt , eigen values will be of murder and assualt 
principal_comp=pd.DataFrame(pca.fit_transform(scale(df)),index=df.index,columns=['PC1','PC2','PC3','PC4'])
print principal_comp.head()
```

```python
pca_load=pd.DataFrame(pca.fit(scale(df)).components_.T,index=df.columns,columns=['v1','v2','v3','v4'])
print pca_load
```
